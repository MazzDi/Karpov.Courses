```python
# В данных имеется как дата аренды, так и её точное время начала и окончания с точностью до секунд. 
# Примените метод pd.resample() и агрегируйте данные по дням. 
# В качестве ответа укажите максимальное число аренд за день.

bikes_Q1.resample(rule='D').count().trip_id.max()
```

```python
# Посмотрим на распределение количества аренд для разных групп пользователей (usertype) — 
# customers и subscribers в данных за апрель. Данные за нужный период можно подгрузить отсюда.

bikes = pd.read_csv('https://stepik.org/media/attachments/lesson/367415/bikes_april.csv', 
                    parse_dates = ['start_time', 'end_time'])
```

```python
bikes_count = bikes.groupby('usertype') \
    .trip_id \
    .resample(rule='D') \
    .count()
```

```python
# Сделайте ресемпл по дням для каждой группы и в качестве ответа укажите число аренд за 18 апреля, 
# сделанных пользователями типа Subscriber.

bikes_count.loc[('Subscriber', '2018-04-18')]
```

```python
# Посмотрим на данные за период с апреля по декабрь.

# Объедините сэмплы данных за нужные месяцы в один общий датасет bikes. 
# Сделайте преобразование по дням для каждой группы пользователей (usertype), затем выберите дни, в 
# которые число аренд, сделанных customers, было больше, чем у subscribers.
```

```python
df_Q2_apr = pd.read_csv('https://stepik.org/media/attachments/lesson/367415/bikes_q2_sample_apr.csv')
df_Q2_may = pd.read_csv('https://stepik.org/media/attachments/lesson/367415/bikes_q2_sample_may.csv')
df_Q2_jun = pd.read_csv('https://stepik.org/media/attachments/lesson/367415/bikes_q2_sample_jun.csv')
df_Q3_jul = pd.read_csv('https://stepik.org/media/attachments/lesson/367415/bikes_q3_sample_july.csv')
df_Q3_aug = pd.read_csv('https://stepik.org/media/attachments/lesson/367415/bikes_q3_sample_aug.csv')
df_Q3_sep = pd.read_csv('https://stepik.org/media/attachments/lesson/367415/bikes_q3_sample_sep.csv')
df_Q4_oct = pd.read_csv('https://stepik.org/media/attachments/lesson/367415/bikes_q4_sample_oct.csv')
df_Q4_nov = pd.read_csv('https://stepik.org/media/attachments/lesson/367415/bikes_q4_sample_nov.csv')
df_Q4_dec = pd.read_csv('https://stepik.org/media/attachments/lesson/367415/bikes_q4_sample_dec.csv')
```

```python
# Объедините сэмплы данных за нужные месяцы в один общий датасет bikes
bikes = pd.concat([df_Q2_apr, df_Q2_may, df_Q2_jun, df_Q3_jul, df_Q3_aug, df_Q3_sep, df_Q4_oct, df_Q4_nov, df_Q4_dec])
```

```python
bikes.start_time = pd.to_datetime(bikes.start_time)
bikes.end_time = pd.to_datetime(bikes.end_time)
```

```python
bikes.dtypes
```

```python
bikes.head()
```

```python
bikes
```

```python
# Сделайте преобразование по дням для каждой группы пользователей (usertype), затем выберите дни, 
# в которые число аренд, сделанных customers, было больше, чем у subscribers.

bikes_cus_sub = bikes.set_index('start_time') \
    .groupby('usertype') \
    .bikeid \
    .resample(rule='D').count() \
    .reset_index() \
    .pivot(index='start_time', columns='usertype', values='bikeid')

bikes_cus_sub.query('Customer > Subscriber')
```

```python
# В переменную bikes_summer сохраните наблюдения с 1 июня по 31 августа. 

bikes_summer = bikes.loc['2018-06-01':'2018-08-31']
```

```python
bikes_summer.head()
```

```python
# Затем запишите в top_destination наиболее популярный пункт назначения (его название).
top_destination = bikes_summer.to_station_name.value_counts().idxmax()
```

```python
# Агрегируйте данные по дням.
bad_day = bikes_summer \
    .query('to_station_name == @top_destination') \
    .resample(rule='D') \
    .trip_id \
    .count() \
    .idxmin()
```

```python
# В какой день в полученный пункт (top_destination) было совершено меньше всего поездок.
bad_day.strftime('%Y-%m-%d')
```

```python
bikes.set_index('start_time', inplace=True)
```

```python
bikes_summer['day'] = bikes_summer.index.day_name()
```

```python
bikes_summer.head()
```

```python
weekdays = ['Friday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday']
```

```python
weekends = ['Saturday', 'Sunday']
```

```python
bikes_summer['is_weekend'] = np.where(bikes_summer.day.isin(weekends), 1, 0)
```

```python
bikes_summer.head()
```

```python
bikes_summer.groupby(['day','is_weekend', 'to_station_name']) \
    .agg({'to_station_name': 'count'}) \
    .rename(columns={'to_station_name': 'count'}) \
    .reset_index() \
    .pivot(index='to_station_name', columns='day', values='count') \
    .idxmax()
```

```python
bikes_summer.query('is_weekend == 0').to_station_name.value_counts()
```
