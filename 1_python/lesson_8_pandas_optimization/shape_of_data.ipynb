```python
import pandas as pd
```

```python
pokemon = pd.read_csv('/home/jupyter-n-korotkov/lesson_8/Pokemon.csv')
pokemon.head()
```

```python
pokemon.columns = map(str.lower, pokemon.columns)
pokemon.columns = pokemon.columns.str.replace(' ', '_').str.replace('.','')
pokemon.rename(columns={'#': 'id'}, inplace = True)
legends = pokemon[['generation', 'legendary']].groupby('generation').legendary.value_counts().to_frame()
```

```python
legends
```

```python
legends.rename(columns={'legendary': 'legendary_count'}, inplace=True)
legends_unstacked = legends.unstack()
```

```python
legends_unstacked
```

```python
pokemon_unstacked = pokemon.groupby(['generation', 'type_1']).legendary.value_counts().unstack(fill_value=0)
```

```python
pokemon_unstacked.idxmax()
```
